Load datasets from CSV

```python
#Loading Dataset
users=pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
sessions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
products =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
transactions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
```

```python
#Descripbe the dataframes
sessions.head()
transactions.head()
products.head()
users.head()
```

12. Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)

```python
#correcting transcation data
transactions.rename(columns={'UserID':'Quantit'},inplace=True)
transactions.rename(columns={'Quantity':'UserID','Quantit':'Quantity'},inplace=True)
transactions.head()
```

```python
#left joining transactions by user id
transactions.merge(users,on='UserID',how='left')
```

13. Which transactions have a UserID not in users?

```python
#Do an anti join to find 
transactions[~transactions['UserID'].isin(users['UserID'])]
```

14. Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)

```python
#Do an Inner Join
transactions.merge(users,on='UserID',how='inner')
```

15. Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)

```python
#Do outter join 
transactions.merge(users,on='UserID',how='outer')
```

16. Determine which sessions occurred on the same day each user registered

```python
#Need to join on two columns 
users.merge(sessions,how='inner',left_on=['UserID', 'Registered'], right_on=['UserID', 'SessionDate'])
```

17. Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
#Find No of unique elements
n_users=users.shape[0]
n_product=products.shape[0]
#Create separate dataframes
df1 = pd.DataFrame({'ID': np.repeat(1, n_users), 'UserID': users.UserID})
df2 = pd.DataFrame({'ID': np.repeat(1, n_product), 'ProductID': products.ProductID})
#Merge the above dataframes
df3=pd.merge(df1, df2,on='ID')[['UserID', 'ProductID']]
df3
```

18. Determine how much quantity of each product was purchased by each user

```python
#Merge two dataframes and calculate te sum of suantities 
pd.merge(df3, transactions, how='left', on=['UserID', 'ProductID']).groupby(['UserID', 'ProductID']).apply(
    lambda x: pd.Series(dict(Quantity=x.Quantity.sum()))).reset_index()
```

19. For each user, get each possible pair of pair transactions (TransactionID1, TransacationID2)

```python
#do a self join
transactions.merge(transactions, on='UserID')
```

20. Join each user to his/her first occuring transaction in the transactions table

```python
#Group transactions
tr1=transactions.groupby('UserID').first()
#Merger transactionswith users
df4=users.merge(tr1, how='left', on='UserID')
df4
```
